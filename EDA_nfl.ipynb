```python
# get the number of missing data points per column
missing_values_count = nfl_data.isnull().sum()

missing_values_count
```

```python
# remove all the rows that contain a missing value
nfl_data.dropna()
```

It seems all the rows has null values. So after removing the null vales, there are no rows in the data

```python
# remove all columns with at least one missing value
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# just how much data did we lose
print("Columns in original dataset: %d \n" % nfl_data.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

#We have lost a lot of data. Another way is the null value imputation.
- <1 % null values - drop the null values
- <7 % null values - impute the null value with central tendency
-10% null values - drop the columns

#Replacement strategy
- continuous data mean, median, linear regression
- Categorical data constant, mode
- Time series data
    - data without trend and seasonality - mean, median, mode, random sample imputation
    - data with trend and without seasonality - Linear interpolation
    - data with trend and seasonality - seasonal adjustment + interpolation
